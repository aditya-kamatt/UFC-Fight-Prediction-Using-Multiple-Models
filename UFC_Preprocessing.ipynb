```python
import pandas as pd
import numpy as np

df = pd.read_csv('fight_data.csv')

relevant_columns = [
    'f_1', 'f_2', 'winner', 'knockdowns', 'total_strikes_att',
    'total_strikes_succ', 'sig_strikes_att', 'sig_strikes_succ',
    'takedown_att', 'takedown_succ', 'submission_att',
    'reversals', 'ctrl_time'
]

df_relevant = df[relevant_columns]

def convert_time_to_seconds(time_str):
    if isinstance(time_str, str):
        minutes, seconds = map(int, time_str.split(":"))
        return minutes * 60 + seconds
    return 0

df_relevant['ctrl_time'] = df_relevant['ctrl_time'].replace(['--', np.nan], '0:00')
df_relevant['ctrl_time'] = df_relevant['ctrl_time'].apply(convert_time_to_seconds)

df_relevant['strikes_att_diff'] = df_relevant['total_strikes_att'].diff()
df_relevant['strikes_succ_diff'] = df_relevant['total_strikes_succ'].diff()
df_relevant['sig_strikes_att_diff'] = df_relevant['sig_strikes_att'].diff()
df_relevant['sig_strikes_succ_diff'] = df_relevant['sig_strikes_succ'].diff()
df_relevant['takedown_att_diff'] = df_relevant['takedown_att'].diff()
df_relevant['takedown_succ_diff'] = df_relevant['takedown_succ'].diff()
df_relevant['submission_att_diff'] = df_relevant['submission_att'].diff()
df_relevant['ctrl_time_diff'] = df_relevant['ctrl_time'].diff()


df_final = df_relevant[['winner', 'strikes_att_diff', 'strikes_succ_diff', 'sig_strikes_att_diff',
                        'sig_strikes_succ_diff', 'takedown_att_diff', 'takedown_succ_diff',
                        'submission_att_diff', 'ctrl_time_diff']]

df_final.dropna(inplace=True)

df_final.to_csv('cleaned_fight_data.csv', index=False)
```
